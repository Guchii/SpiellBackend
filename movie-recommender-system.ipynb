We have to flatten the genres column above

```python
from ast import literal_eval # for parsing the strings in columns
```

```python
def convert(obj):
    genres = []
    for i in literal_eval(obj):
        genres.append(i['name'])
    return genres
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

done üëç
Now same for keywords

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_cast(obj):
    cast = []
    for idx, i in enumerate(literal_eval(obj)):
        if idx < 3:
            cast.append(i['name'])
        else:
            break
    return cast
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(1)
```

resolving crew list

```python
# we have to fetch only directors
def convert_crew(obj):
    dire = []
    for idx, i in enumerate(literal_eval(obj)):
        if i['job'] == "Director":
            dire.append(i['name'])
            break
    return dire
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head(1)
```

done üëç

convert overview from a big list to a list of strings

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

Now we'll join names together, 
'Sam Worthington' => 'SamWorthington'
so that model can distinguish between sam worthington and sam mendes

```python
split_n_join = lambda x: ["".join(i.split()) for i in x]
# movies['genres'] = movies['genres'].apply(split_n_join)
for col in ['genres', 'keywords', 'cast', 'crew']:
    movies[col] = movies[col].apply(split_n_join)
```

```python
movies.head(1)
```

done üëç

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

## Vectorisation
- Essential to work with string data
- in order to calculate similarity b/w movie tags
- We'll not use stop words `are and if to from`
- `text => text vectors` using **BAG OF WORDS**
- find closest vectors during prediction


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
my_cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = my_cv.fit_transform(new_df['tags']).toarray()
# scipy sparse matrix to numpy array (.toarray())
vectors
```

```python
vectors.shape
```

```python
# my_cv.get_feature_names() returns the 5000 most common words
len(my_cv.get_feature_names())
```

## Stemming
to make words like 'activities' and 'activity' same
extracting root words

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
for i in ["love", "loving", "loved"]:
    print(ps.stem(i))
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# scipy sparse matrix to numpy array (.toarray())
vectors = my_cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
# my_cv.get_feature_names() returns the 5000 most common words
my_cv.get_feature_names()
```

### calculate distance b/w vectors 
- lesser the distance more the similarity score
- we'll calculate the angle b/w the vectors (cosine distance) not the euclidean distance since we are working with **high dimensional data**

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

the above cell calculates the cosine distance b/w each movie and creates a cosine matrix

```python
similarity[0]
```

```python
def recommend(movie):
    idx = None
    for i, mov in enumerate(list(new_df['title'])):
        if movie.lower() == mov.lower():
           idx = i 
           break
    distances = similarity[idx]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("aliens")
```
